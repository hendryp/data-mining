```python
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
covidDataset = pd.read_csv('covid.csv',delimiter=',',header=0)
```

```python
covidDataset.head()
```

```python
covidDataset["Region"] = pd.factorize(covidDataset.Region)[0]
```

```python
covidDataset.head()
```

```python
covidDataset = covidDataset.drop(labels="Id",axis=1)
```

```python
print(covidDataset)
```

```python
covidDataset = covidDataset.to_numpy()
```

```python
print(covidDataset)
```

```python
dataTraining = np.concatenate((covidDataset[0:40, :], covidDataset[50:90, :]),axis=0)
dataTesting = np.concatenate((covidDataset[40:50, :], covidDataset[90:100, :]),axis=0)
```

```python
print(dataTesting)
len(dataTesting)
```

```python
print(dataTraining)
len(dataTraining)
```

```python
inputTraining = dataTraining[:,0:4]
inputTesting = dataTesting[:,0:4]
labelTraining = dataTraining[:,4]
labelTesting = dataTesting[:,4]
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model = model.fit(inputTraining, labelTraining)
```

```python
hasilPrediksi = model.predict(inputTesting)
```

```python
print("Label sebenarnya ",labelTesting)
print("Hasil prediksi ",hasilPrediksi)
```

```python
prediksiBenar = (hasilPrediksi == labelTesting).sum()
prediksiSalah = (hasilPrediksi != labelTesting).sum()
```

```python
print("Prediksi benar: ", prediksiBenar, " data")
print("Prediksi salah: ", prediksiSalah, " data")
print("Akurasi: ",prediksiBenar/(prediksiBenar+prediksiSalah)*100,"%")
```
